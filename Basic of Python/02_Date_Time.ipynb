```python
import pandas as pd
```

```python
import datetime as dt
date = dt.datetime.now()
print(date)
```

```python
print(date.year)
```

```python
print(date.month)
```

```python
print(date.strftime("%B"))
```

```python
# "%A" - use for day
# "%B" - use for month

print("Day :" + date.strftime("%A"))
print("Month : "+ date.strftime("%B"))
print("Date : " + date.strftime("%C"))
print("Month/Date/Year : " + date.strftime("%D"))
print("Year-Month-Date : " + date.strftime("%F"))
print("Only Year : " + date.strftime("%G"))
print("Only Hours in 24 format : " + date.strftime("%H"))
print("Only Minutes  : " + date.strftime("%M"))
print("Indian format : " + date.strftime("%c"))

```
